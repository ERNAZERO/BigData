<h1>После работы с разными моделями регрессии, я для себя подчеркнул несколько заметок.</h1>

<h2>Первая модель регрессии была - KNeighborsRegressor.</h2>
Эта модель хорошо предсказывает значения на основе близости соседних точек, но может быть нестабильной из-за выбросов.
<h2>Вторая модель с которой я работал была - LinearRegression.</h2>
Простая модель, предсказывающая значения, когда признаки линейно связаны, но если признаки сильно похожи друг на друга, это может повлиять на точность предсказаний.
<h2>Гребневая регрессия (Ridge) и Lasso Regression - эти модели помогают предотвратить переобучение</h2>. Ridge сохраняет все признаки, но уменьшает их веса, в то время как Lasso может убрать несколько ненужных признаков.
<h2>И наконец - LogisticRegression.</h2> Хорошо работает для предсказания категорий. Но лучше использовать для бинарных классификаций.
В целом, для себя я понял, что разные модели регрессии имеют свои преимущества и ограничения. Выбор модели зависит от типа данных и задачи. Важно настраивать параметры моделей для лучших результатов, которые мы хотим получить.
