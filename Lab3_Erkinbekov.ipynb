4.	Вывести графики п.2 и 3 в разные поля при помощи функции subplot()

```python
x_quadratic = np.linspace(-10, 10, 400)
y_quadratic = a * x_quadratic**2 + b * x_quadratic + c

x_harmonic = np.linspace(0, 2 * np.pi, 400)
y_harmonic = np.sin(x_harmonic)

plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plt.plot(x_quadratic, y_quadratic, label=f'{a}x^2 + {b}x + {c}')
plt.xlabel('X-ось')
plt.ylabel('Y-ось')
plt.title('График квадратичной функции')
plt.grid(True)
plt.legend()

plt.subplot(1, 2, 2)
plt.plot(x_harmonic, y_harmonic, label='sin(x)')
plt.xlabel('X-ось')
plt.ylabel('Y-ось')
plt.title('График гармонической функции')
plt.grid(True)
plt.legend()

plt.tight_layout()
plt.show()
```

5.	Построить графики п.2 и 3 в одной координатной плоскости, сделать подписи графиков, указать легенду.

```python
x_quadratic = np.linspace(-2, 2, 400)
y_quadratic = a * x_quadratic**2 + b * x_quadratic + c

x_harmonic = np.linspace(0, 2 * np.pi, 400)
y_harmonic = np.sin(2 * x_harmonic)

plt.figure(figsize=(12, 6))

plt.plot(x_quadratic, y_quadratic, label=f'{a}x^2 + {b}x + {c}', color='blue')
plt.xlabel('X-ось')
plt.ylabel('Квадратичная функция', color='blue')
plt.grid(True)
plt.legend(loc='upper left')

ax2 = plt.gca().twinx()

ax2.plot(x_harmonic, y_harmonic, label='sin(2x)', color='red')
ax2.set_ylabel('Синусоида', color='red')
ax2.legend(loc='upper right')

plt.title('Графики квадратичной и гармонической функций')
plt.tight_layout()
plt.show()
```

9.	Построить график произвольной каркасной поверхности.

```python
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111, projection='3d')

u = np.linspace(0, 2 * np.pi, 100)
v = np.linspace(0, np.pi, 100)
x = 5 * np.outer(np.cos(u), np.sin(v))
y = 5 * np.outer(np.sin(u), np.sin(v))
z = 5 * np.outer(np.ones(np.size(u)), np.cos(v))

ax.plot_surface(x, y, z, color='none', edgecolor='b')

ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_title('Произвольная каркасная поверхность')

plt.show()
```
